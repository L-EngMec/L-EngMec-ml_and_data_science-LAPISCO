# 5

Crie 3 vetores de atributos distintos, e utilizando as distâncias euclidiana e Manhattan, determine a qual classe cada vetor de atributos pertence.

```python
# Vetores
v1 = [4.2, 3.6, 0.3]
v2 = [5.6, 2.2, 6.6]
v3 = [4.3, 2.0, 0.3]

classe1 = [5.00600, 3.428000, 1.462000, 'setosa']
classe2 = [5.936000, 2.770000, 4.260000,'Versicolor']
classe3 = [6.58800, 2.974000, 5.552000, 'Virginica']
```

Os valores escolhidos para cada classe são oriundos dos valores médios de cada classe presente no dataframe. Então para os elementos do dataframe com a classe 'setosa', os valores presentes na classe1 são os valores médios dos elementos categorizados como 'setosa'.

```python
def lista_com_menores_elementos_entre_tres(lista1, lista2, lista3):
    elementos_menores = [min(val1, val2, val3) for val1, val2, val3 in zip(lista1, lista2, lista3)]
    
    if elementos_menores == lista1:
        return lista1
    elif elementos_menores == lista2:
        return lista2
    else:
        return lista3

def dist_euclidiana_3D(V1, V2):
    v1 = V1[0:3] # garantindo que ambos vão ter as mesmas dimensões
    v2 = V2[0:3] # garantindo que ambos vão ter as mesmas dimensões
    parcelas = []
    for i in range(len(v1)):
        subtracao = v1[i] - v2[i]
        
        if subtracao < 0:
            subtracao = subtracao*(-1)
        parcelas.append(subtracao)
    distancia = (parcelas[0]**2 + parcelas[1]**2 + parcelas[2]**2)**(1/2)
    
    return distancia

def dist_manhattan_3D(V1, V2):
    v1 = V1[0:3] # garantindo que ambos vão ter as mesmas dimensões
    v2 = V2[0:3] # garantindo que ambos vão ter as mesmas dimensões
    parcelas = []
    for i in range(len(v1)):
        subtracao = v1[i] - v2[i]
        
        if subtracao < 0:
            subtracao = subtracao*(-1)
        parcelas.append(subtracao)
    return parcelas

def classificacao(vetor, distancia):
    classe1 = [5.00600, 3.428000, 1.462000, 'setosa']
    classe2 = [5.936000, 2.770000, 4.260000,'Versicolor']
    classe3 = [6.58800, 2.974000, 5.552000, 'Virginica']
    
    if distancia == 'euclidiana':
        d1 = dist_euclidiana_3D(vetor, classe1)
        d2 = dist_euclidiana_3D(vetor, classe2)
        d3 = dist_euclidiana_3D(vetor, classe3)
        menor_distancia = min(d1, d2, d3)
        if menor_distancia == d1:
            return classe1[3]
        elif menor_distancia == d2:
            return classe2[3]
        else:
            return classe3[3]
    elif distancia == 'manhattan':
        d1 = dist_manhattan_3D(vetor, classe1)
        d2 = dist_manhattan_3D(vetor, classe2)
        d3 = dist_manhattan_3D(vetor, classe3)
        menor_distancia = lista_com_menores_elementos_entre_tres(d1, d2, d3)
        if menor_distancia == d1:
            return classe1[3]
        elif menor_distancia == d2:
            return classe2[3]
        else:
            return classe3[3]
```

```python
# Usando distancia Euclidiana
print(classificacao(v1, 'euclidiana'))
print(classificacao(v2, 'euclidiana'))
print(classificacao(v3, 'euclidiana'))
```

```python
# Usando distancia Manhattan
print(classificacao(v1, 'manhattan'))
print(classificacao(v2, 'manhattan'))
print(classificacao(v3, 'manhattan'))
```
