```python
import pandas as pd

path = '/home/levy_silveira/Downloads/Dados/Preço-Celular/train (1).csv'
data = pd.read_csv(path)

x = data.drop(columns=['price_range'])
y = data['price_range']

data
```

# 2
Utilize a técnica k-fold cross validation. Compare os resultados com o melhor resultado da Atividade 6.

```python
def average(values):
    sum = 0
    for item in values:
        sum += item
    return sum/len(values)
```

```python
from sklearn.neighbors import KNeighborsClassifier 
from sklearn.model_selection import cross_val_score

knn = KNeighborsClassifier()
accuracies = cross_val_score(knn, X=x, y=y, cv=10, scoring='accuracy')

Average = average(accuracies)

print(f'Accuracies: {accuracies};\nAverage: {Average}')
```

Realize experimentos utilizando normalização e padronização. Compare os resultados com o melhor resultado da Atividade 6.

## Normalização

```python
from sklearn.preprocessing import MinMaxScaler
normalizador = MinMaxScaler()
normalizador.fit(data)
dados_normalizados = pd.DataFrame(normalizador.transform(data), columns = data.columns)
dados_normalizados.head()
```

```python
x_normalizado = dados_normalizados.drop(columns='price_range')
y_normalizado = dados_normalizados['price_range']
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(x_normalizado, y, test_size=0.2, random_state=42)
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score

lr = LogisticRegression()
lr.fit(X_train, y_train)
pred = lr.predict(X_test)
```

```python
from sklearn import metrics
from sklearn.metrics import accuracy_score

confusion_matrix = metrics.confusion_matrix(y_test, pred)
acc = accuracy_score(y_test, pred)

print("Acc: ", acc)
```

## Padronização
